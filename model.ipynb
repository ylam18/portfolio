```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('turkeyEarthquakes.csv')
depth = dataframe['Derinlik'].values.reshape(-1,1)
magnitude = dataframe['xM'].values
```

```python
from sklearn.model_selection import train_test_split
depth_train, depth_test, magnitude_train, magnitude_test = train_test_split(depth, magnitude, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
model = LinearRegression()
model.fit(depth_train, magnitude_train)
magnitude_pred = model.predict(depth_test)
MSE = mean_squared_error(magnitude_test, magnitude_pred)
r_score = r2_score(magnitude_test, magnitude_pred)
print("The Mean Squared Error is: ", MSE)
print("The R-squared value is: ", r_score)
```

```python
plt.scatter(depth_train, magnitude_train, color = 'b')
plt.plot(depth_test, magnitude_pred, color = 'g')
plt.show()
```
